### One-Day SPX Returns

```python
# load Data
# SPX 
SPX_DailyReturn = pd.read_csv("./Data/SPX_Daily.csv")
SPX_DailyReturn.rename(columns={'sprtrn':'SPX'}, inplace=True)
SPX_DailyReturn['SPX'] = SPX_DailyReturn['SPX']*100
SPX_DailyReturn['DATE'] = pd.to_datetime(SPX_DailyReturn['DATE']).dt.date
# VIX
VIX_DailyReturn = pd.read_csv('./Data/VIX_Daily.csv')
VIX_DailyReturn.rename(columns={'Date':'DATE', 'vix':'VIX'}, inplace=True)
VIX_DailyReturn['DATE'] = pd.to_datetime(VIX_DailyReturn['DATE']).dt.date
VIX_DailyReturn.sort_values(by = 'DATE')
# merge data
SPX_DailyData = pd.merge(SPX_DailyReturn, VIX_DailyReturn, on='DATE', how='left')
SPX_DailyData
# Top worst
Top10Worst = SPX_DailyData.nsmallest(10, 'SPX').reset_index(drop=True)
Top10Worst.insert(0, 'Rank', range(1,11))
Top10Worst[['SPX', 'VIX']] = Top10Worst[['SPX', 'VIX']].round(2).fillna("")
```

```python
# 使用 tabulate 生成三线表
table_str = tabulate(Top10Worst, headers="keys", tablefmt="pipe", showindex=False)
print(table_str)
```

```python
# plot
plt.figure(figsize=(5, 5))
plt.plot(SPX_DailyReturn['DATE'], SPX_DailyReturn['SPX'], 
         linestyle='None', color='blue', marker='o', markersize=1)
plt.ylim(-25, 15)
plt.scatter(Top10Worst['DATE'], Top10Worst['SPX'], facecolors='none', edgecolors='red', 
            marker='o', s=50, linewidths=0.5)
plt.title('Daily SPX Returns (%)')
plt.show()
```

## Class 2

### Annual Stock Returns

```python
# data
CRSP_VMP_Annual = pd.read_csv("./Data/CRSP_VMP_Annual.csv")
CRSP_VMP_Annual.rename(columns={'caldt':'Year', 'vwretd':'Ret'}, inplace=True)
CRSP_VMP_Annual['Ret'] = CRSP_VMP_Annual['Ret']*100
CRSP_VMP_Annual['Year'] = pd.to_datetime(CRSP_VMP_Annual['Year']).dt.year
TopWorst = CRSP_VMP_Annual.nsmallest(5, 'Ret').reset_index(drop=True)
TopBest = CRSP_VMP_Annual.nlargest(7, 'Ret').reset_index(drop=True)
# plot
plt.figure(figsize=(5,5))
plt.plot(CRSP_VMP_Annual['Year'], CRSP_VMP_Annual['Ret'], linewidth=1, color='blue', marker="*", markersize=3, linestyle='-')
plt.scatter(TopBest['Year'], TopBest['Ret'], facecolors='none', edgecolors='green', marker='o', s=50)
for i in range(len(TopBest)):
    plt.text(TopBest['Year'][i]+2, TopBest['Ret'][i]-1, f"{TopBest['Ret'][i]:.0f}%", fontsize=6, color='green', fontweight='bold', ha='left')
plt.scatter(TopWorst['Year'], TopWorst['Ret'], facecolors='none', edgecolors='red', marker='o', s=50)
for i in range(len(TopWorst)):
    plt.text(TopWorst['Year'][i]+2, TopWorst['Ret'][i]-1, f"-{TopWorst['Ret'][i]:.0f}%", fontsize=6, color='red', fontweight='bold', ha='left')
plt.ylim(-50, 70)
plt.yticks(fontsize=8)
plt.xticks([1940, 1960, 1980, 2000], fontsize=8)
plt.tick_params(axis='both', which='major', direction='in', length=6, width=1)
plt.title('Annual Stock Returns (in Percent) from 1927 through 2015', fontsize=8, fontweight='bold')
plt.show()
```

```python
mu = CRSP_VMP_Annual['Ret'].mean().round(0) 
sd = round(CRSP_VMP_Annual['Ret'].std(),0)
N = CRSP_VMP_Annual.shape[0]
print(f"the sample mean is {mu}%")
print(f"the standard deviation of mu is {sd}%")
print(f"the number of observations is {N}")
print(f"the standard error of mu_hat is {round(sd/np.sqrt(N-1),2)}%")
```

```python
plt.figure(figsize=(6,4))
plt.hist(CRSP_VMP_Annual['Ret'], bins=10, color='darkblue', rwidth=0.8)
plt.ylim(0, 18)
plt.xlim(-60, 80)
plt.annotate('', xy=(-40, 3.0), xytext=(-40, 6), arrowprops=dict(facecolor='black', arrowstyle="->"))
plt.text(-55, 6.5, "1931:-44%\n2008:-37%\n1937:-35%", fontsize=8)
plt.annotate('', xy=(-30, 2.0), xytext=(-30, 4), arrowprops=dict(facecolor='black', arrowstyle="->"))
plt.text(-35, 4, "1930:-49%\n1974:-28%", fontsize=8)
plt.annotate('', xy=(40, 5.0), xytext=(40, 8), arrowprops=dict(facecolor='black', arrowstyle="->"))
plt.text(40, 8, "1935:45%\n1958:45%\n1928:39%\n1945:39%\n1975:38%", fontsize=8)
plt.annotate('', xy=(50, 2.0), xytext=(50, 4), arrowprops=dict(facecolor='black', arrowstyle="->"))
plt.text(50, 4, "1933:57%\n1954:50%", fontsize=8)
plt.ylabel("Number of Occurrence", fontsize=8)
plt.xlabel("Scenarios of Possible Annual Returns", fontsize=8)
plt.title("Learning from History: Possible Events and Their Occurence", fontsize=10)
plt.show();
```

```python
plt.figure(figsize=(5,5))
plt.plot(np.linspace(-40,50,1000), norm.pdf(np.linspace(-40,50,1000), 12, 20), 
        color='blue', linewidth=0.5)
plt.plot(np.linspace(-40,50,1000), norm.pdf(np.linspace(-40,50,1000), 12, 2.13), 
        color='red', linewidth=0.5)
plt.fill_between(np.linspace(-40,7.8,500), norm.pdf(np.linspace(-40,7.8,500),12,2.13), color='yellow')
plt.fill_between(np.linspace(16.2,50.8,500), norm.pdf(np.linspace(16.2,50,500),12,2.13), color='yellow')
plt.plot([7.8,7.8], [0,0.025], color='black', linewidth=0.5)
plt.plot([16.2,16.2], [0,0.025], color='black', linewidth=0.5)
plt.annotate(r'pdf of $\hat{\mu}$', xy=(16, 0.16), xytext=(22,0.16), arrowprops=dict(arrowstyle="->"), fontsize=10)
plt.annotate(r'pdf of $R_t$', xy=(-10, 0.02), xytext=(-28,0.03), arrowprops=dict(arrowstyle="->"), fontsize=10)
plt.text(-30,0.16, r"$\hat{\mu}=\frac{1}{N}\sum_{t=1}^NR_t$", fontsize=10)
plt.ylim(0,0.2)
plt.xlim(-40, 50)
plt.xticks(np.arange(-40, 60, 10), fontsize=6)
plt.yticks(np.linspace(0,0.2,11), [f"{i:.0%}" for i in np.linspace(0,0.2,11)], fontsize=6)
plt.xlabel("Return (%)", fontsize=8)
plt.ylabel("Probability Density", fontsize=8)
plt.title("The 95-Percent Confidence Interval", fontsize=10)
plt.show();
```

```python
CRSP_VMP_Month = pd.read_csv("./Data/CRSP_VMP_Monthly.csv")
CRSP_VMP_Month.rename(columns={'DATE':'YearMonth', 'vwretd':'Ret'}, inplace=True)
CRSP_VMP_Month['Ret'] = CRSP_VMP_Month['Ret']*100
CRSP_VMP_Month['YearMonth'] = pd.to_datetime(CRSP_VMP_Month['YearMonth'])
CRSP_VMP_Month = CRSP_VMP_Month[CRSP_VMP_Month['YearMonth'] < "2011-12-31"]
mu_month = CRSP_VMP_Month['Ret'].mean()
N_month = CRSP_VMP_Month.shape[0]
std_month = CRSP_VMP_Month['Ret'].std()
print(f"the mean is {mu_month:.2f}%")
print(f"the number of month is {N_month}")
print(f"the std is {std_month:.2f}%")
```

```python
CRSP_VMP_Month['YearMonth'] = pd.to_datetime(CRSP_VMP_Month['YearMonth'])
plt.figure(figsize=(6,4))
plt.plot(CRSP_VMP_Month['YearMonth'], CRSP_VMP_Month['Ret'], color='blue', linewidth=0.5)
plt.text(CRSP_VMP_Month.iloc[800,0], 30, "mean=0.91%", fontsize=8)
plt.text(CRSP_VMP_Month.iloc[800,0], 25, r"std=5.46%", fontsize=8)
plt.title("Monthly Aggregate Stock Returns (in percent) from Jan 1927 through Dec 2011", fontsize=8)
plt.xticks([pd.Timestamp(f"{year}-01-31") for year in range(1920, 2021, 20)], 
           [str(year) for year in range(1920, 2021, 20)],fontsize=8)
plt.yticks(fontsize=8)
plt.show();
```
